## 1. Introduction

`DICOM Import` in SPM 5 is a conversion tool to convert files with dicom format (output of Siemens scanners) to Neuro Informatics Technology Initiative (NIFTI) format with .nii extension.

files with NIFTI (.nii extension) format can be prosessed with different fMRI analysis software and packages.

## Note:

to use SPM 5, you need to have older versions of MATLAB. For example, here I am using MATLAB2010b version. In newer versions the SPM 5 toolbox is not working. For example I tested the SPM 5 toolbox in MATLAB2015 and `DICOM Import` ended up with an error in command window and did not work.

## 2. Download DICOM file samples

To start work with dicom import section, you need to have some DICOM file samples. I could find a good and small sample of DICOM files here, https://zenodo.org/records/16956.
Also I put those files in the Data/ folder in my github.

## 3. Open DICOM Import

To open DICOM Import section from SPM, open SPM in MATLAB2010 with the following command:`spm`. After this the main window of SPM 5 will apear as follows:

In the main window of SPM 5, click DICOM Import, circled in red, and the following windows will appear:



DICOM Import Window





In the DICOM import (DICOM files), using section 1, select the drive that contains DICOM files, using section 2, and .. available inside it, navigate to the folder that contains DICOM files, for example in my system the files I downloaded from section 2. Download DICOM file samples, I put the files in the directory shown in the above text box called, Dir.

After reaching to the files, in the section 3, you can see the list of DICOM files then right-click in the empty space of section 3 and choose `Select All`, then, list of selected files will be shown in the empty list in the below of the window and then using section 4, click on `Done`.

After clicking `Done`, the next window will appear and will ask you about the output directory that you wish to save there:

In this window, `Output Directory`, using `Drive` and section 1, navigate to the folder you want to save the outputs (just like the previous step), then after seeing final folder in the section 2 list, here `dicomoutput\`, click it in the section 2, then select `Done`. 

Then, the `DICOM Import: setup` window will appear, from the `Output image format...` select `Single file (nii) NIfTI` 

Then, in the next window, from the `Use ICEDims` in filename..., select `No`.

Then the following window will appear and the process of the input DICOM files start. 

When the process finished, if you navigate to the output directory, you can see the .nii and header files processed from the input files there.
