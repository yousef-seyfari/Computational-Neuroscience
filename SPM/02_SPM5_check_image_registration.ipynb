# Image Registration in fMRI (Functional Magnetic Resonance Imaging):

Image registration in fMRI is a crucial process that aligns and matches different brain scans or images from the same individual. This alignment ensures accurate comparisons and analyses of brain activity over time or across different conditions.

For checking image registration, there are two methods: GUI and command line.

## 1. GUI method:

1.1 Step 1:

    -Click the Check Reg button (Green highlighted).
    -The Images to Display dialogue will appear.

In the Images to Display dialogue, navigate to the NIFTI format images (.nii extensions).

Note: .nii files are created using [SPM->DICOM Import](01_SPM5_dicom_import.ipynb).

Using the left box (highlighted green), navigate to the path containing .nii file (e.g., dicomoutput3 directory). After identifying the target folder, single click on it in the left box. Its contents will be shown on the right box. When you see your desired .nii file on the right box, click on it to select it and show it in the lower box. Click Done. Subsequently, the Graphics window will display your images for you to check the image registration.

## 2. Command Line

In the command line, type spm_check_registration, and the same Images to Display dialogue will appear, but this time its name is Select images.

In order to pass the path to the NIFTI format images as a parameter, use the following example:
`spm_check_registration('I:/Research_Active/Neuro/Data/dicomoutput3/sRJN7270540X-0201-00002-000001-01.nii')`
